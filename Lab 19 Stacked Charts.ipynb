### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}

df['AgeNumeric'] = df['Age'].map(age_mapping)


median_scores =df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot stacked bar chart
median_scores.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)

plt.title("Stacked Chart of Median Job Satisfaction (Points 6 & 7) Across Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median Job Satisfaction Score")
plt.legend(["JobSatPoints_6", "JobSatPoints_7"])
plt.tight_layout()
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
df_clean = df.dropna(subset=['Employment', 'JobSatPoints_6', 'JobSatPoints_7'])

top_employment = df_clean['Employment'].value_counts().head(10).index
df_top = df_clean[df_clean['Employment'].isin(top_employment)]

# Group by Employment and calculate median for the two satisfaction points
median_scores = df_top.groupby('Employment')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot stacked bar chart
median_scores.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])

plt.title("Stacked Chart of Job Satisfaction (Points 6 & 7) Across Top 10 Employment Statuses")
plt.xlabel("Employment Status")
plt.ylabel("Median Job Satisfaction Score")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
age_filtered = df[(df['AgeNumeric'] >= 25) & (df['AgeNumeric'] <= 34)]
median_comp = age_filtered['ConvertedCompYearly'].median(skipna=True)

if pd.notna(median_comp):
    plt.bar(['Age 30-35'], [median_comp], color='skyblue')
    plt.ylabel('Median Compensation (USD)')
    plt.title('Median Compensation for Age 30-35')
    plt.show()
else:
    print("No compensation data.")
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Drop rows where both values are missing
df_clean = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Age'])

# Group by Age and calculate median compensation and median job satisfaction
median_scores = df_clean.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# Plot stacked bar chart
median_scores.plot(kind='bar', stacked=True, figsize=(10,6), color=['skyblue', 'orange'])

plt.title("Stacked Chart of Median Compensation and Job Satisfaction Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Median Value")
plt.xticks(rotation=45, ha='right')
plt.legend(["Median Compensation", "Median Job Satisfaction"])
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Drop rows with missing age or database info
df_clean = df.dropna(subset=['Age', 'DatabaseWantToWorkWith'])

# Map Age to numeric midpoints if needed
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}
df_clean['AgeNumeric'] = df_clean['Age'].map(age_mapping)

# Split multiple databases listed per respondent
df_expanded = df_clean.assign(Database=df_clean['DatabaseWantToWorkWith'].str.split(';')).explode('Database')

# Count frequency of each database per age group
db_counts = df_expanded.groupby(['Age', 'Database']).size().unstack(fill_value=0)

# Keep top 5 databases overall
top_dbs = db_counts.sum().sort_values(ascending=False).head(5).index
db_counts_top = db_counts[top_dbs]

# Convert counts to proportions per age group
db_props = db_counts_top.div(db_counts_top.sum(axis=1), axis=0)

# Plot stacked bar chart
db_props.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')

plt.title("Proportion of Top 5 Desired Databases by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Proportion of Respondents")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Database")
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
df_clean = df.dropna(subset=['Employment', 'JobSatPoints_6']).copy()

df_clean['JobSatLevel'] = df_clean['JobSatPoints_6'].round().astype(int)

top_employment = df_clean['Employment'].value_counts().head(10).index
df_clean['EmploymentTop'] = df_clean['Employment'].where(df_clean['Employment'].isin(top_employment), 'Other')

employment_counts = df_clean.groupby(['JobSatLevel', 'EmploymentTop']).size().unstack(fill_value=0)

employment_counts.plot(kind='bar', stacked=True, figsize=(12,6), colormap='tab20')

plt.title("Distribution of Top 10 Employment Types Across Job Satisfaction Levels")
plt.xlabel("Job Satisfaction Level (Rounded)")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=0)
plt.legend(title="Employment Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}
df_clean['AgeNumeric'] = df_clean['Age'].map(age_mapping)

df_expanded = df_clean.assign(Language=df_clean['LanguageAdmired'].str.split(';')).explode('Language')

# Count frequency of each language per age group
lang_counts = df_expanded.groupby(['Age', 'Language']).size().unstack(fill_value=0)

top_languages = lang_counts.sum().sort_values(ascending=False).head(5).index
lang_counts_top = lang_counts[top_languages]

lang_props = lang_counts_top.div(lang_counts_top.sum(axis=1), axis=0)

lang_props.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')

plt.title("Proportion of Top 5 Admired Programming Languages by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Proportion of Respondents")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Programming Language")
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python


df_clean = df.dropna(subset=['Employment', 'PlatformAdmired']).copy()
df_expanded = df_clean.assign(Platform=df_clean['PlatformAdmired'].str.split(';')).explode('Platform')

platform_counts = df_expanded.groupby(['Employment', 'Platform']).size().unstack(fill_value=0)
platform_props = platform_counts.div(platform_counts.sum(axis=1), axis=0).head(5)



platform_props.plot(kind='bar', stacked=True, figsize=(14,7), colormap='tab20')

plt.title("Proportion of Admired Platforms by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Proportion of Respondents")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
