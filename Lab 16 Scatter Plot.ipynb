#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
df = pd.read_csv(file_path)

# Convert Age to numeric using midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert JobSatPoints_6 to numeric
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['AgeNumeric', 'JobSatPoints_6'])

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['AgeNumeric'], df['JobSatPoints_6'], alpha=0.6, color='blue')
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Age vs Job Satisfaction', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

# Remove extreme outliers for compensation
df = df[df['ConvertedCompYearly'] > 0]
df = df[df['ConvertedCompYearly'] < 500000]  # optional: remove extreme outliers

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['ConvertedCompYearly'], df['JobSatPoints_6'], alpha=0.6, color='green')

plt.xlabel('Yearly Compensation (USD)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Yearly Compensation vs Job Satisfaction', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
# Convert Age to numeric midpoints
import numpy as np
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert JobSatPoints_6 to numeric
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['AgeNumeric', 'JobSatPoints_6'])

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['AgeNumeric'], df['JobSatPoints_6'], alpha=0.6, color='blue', label='Data Points')

# Fit linear regression line
x = df['AgeNumeric']
y = df['JobSatPoints_6']
slope, intercept = np.polyfit(x, y, 1)
plt.plot(x, slope*x + intercept, color='red', linewidth=2, label='Trend Line')

plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Age vs Job Satisfaction with Trend Line', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
df = pd.read_csv(file_path)

# Convert Age to numeric midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert YearsCodePro to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['AgeNumeric', 'YearsCodePro'])
df = df[df['YearsCodePro'] >= 0]  # remove negative values if any

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['AgeNumeric'], df['YearsCodePro'], alpha=0.6, color='purple')

plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Years of Professional Coding Experience', fontsize=12)
plt.title('Age vs Years of Professional Coding Experience', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(file_path)

# Convert Age to numeric midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert other columns to numeric
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['AgeNumeric', 'ConvertedCompYearly', 'JobSatPoints_6'])

# Remove extreme outliers for better visualization
df = df[(df['ConvertedCompYearly'] > 0) & (df['ConvertedCompYearly'] < 500000)]
df = df[df['AgeNumeric'] > 0]

# Bubble sizes proportional to Age
bubble_size = df['AgeNumeric'] * 5  # scale factor for visibility

# Plot bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['ConvertedCompYearly'], df['JobSatPoints_6'], s=bubble_size, alpha=0.6, c=df['AgeNumeric'], cmap='viridis', edgecolor='k')

plt.xlabel('Yearly Compensation (USD)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Compensation vs Job Satisfaction (Bubble Size = Age)', fontsize=14)
cbar = plt.colorbar()
cbar.set_label('Age (years)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')
df = df.dropna(subset=['JobSatPoints_6', 'LanguageHaveWorkedWith'])

# Split multiple languages into separate rows
df_lang = df.assign(LanguageHaveWorkedWith=df['LanguageHaveWorkedWith'].str.split(';')).explode('LanguageHaveWorkedWith')

# Optional: filter top 10 most popular languages
top_languages = df_lang['LanguageHaveWorkedWith'].value_counts().head(10).index.tolist()
df_lang_top = df_lang[df_lang['LanguageHaveWorkedWith'].isin(top_languages)]

# Scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(df_lang_top['LanguageHaveWorkedWith'], df_lang_top['JobSatPoints_6'], alpha=0.6, color='teal')

plt.xlabel('Programming Language', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Job Satisfaction Across Popular Programming Languages', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Employment'])

# Remove extreme outliers for better visualization
df = df[(df['ConvertedCompYearly'] > 0) & (df['ConvertedCompYearly'] < 500000)]

# Unique employment types
employment_types = df['Employment'].unique()
colors = plt.cm.tab10.colors  # use tab10 colormap for up to 10 categories

plt.figure(figsize=(12, 6))

for i, emp in enumerate(employment_types):
    emp_data = df[df['Employment'] == emp]
    plt.scatter(emp_data['ConvertedCompYearly'], emp_data['JobSatPoints_6'],
                alpha=0.6, label=emp, color=colors[i % len(colors)])

plt.xlabel('Yearly Compensation (USD)', fontsize=5)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Compensation vs Job Satisfaction by Employment Type', fontsize=14)
plt.legend(title='Employment Type', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(file_path)

# Convert Age to numeric midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert YearsCodePro to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['YearsCodePro', 'AgeNumeric', 'Country'])
df = df[df['YearsCodePro'] >= 0]

# Optional: only top 5 countries for readability
top_countries = df['Country'].value_counts().head(5).index.tolist()
df_top = df[df['Country'].isin(top_countries)]

# Scatter plot
plt.figure(figsize=(12, 6))
for i, country in enumerate(top_countries):
    subset = df_top[df_top['Country'] == country]
    plt.scatter(subset['AgeNumeric'], subset['YearsCodePro'], alpha=0.6, label=country)

plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Years of Professional Coding Experience', fontsize=12)
plt.title('Years of Coding Experience by Age and Country', fontsize=14)
plt.legend(title='Country')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
